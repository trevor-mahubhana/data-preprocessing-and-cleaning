```python
# Make the changes permanent
customer_df = customer_df.loc[~customer_df.duplicated(subset=['First_name','Surname','Phone_number'])]
```

```python
# View the 'Address' column 
# Surprisingly the address column contains street name,location and city/town
# Need to split it
customer_df.Address
```

```python
# View rows with missing address
customer_df.loc[customer_df.Address.isnull()]
```

```python
# Drop rows with missing address since address is a priority for promotion gift delivery 
# Make the changes permanent
customer_df = customer_df.loc[~customer_df.Address.isnull()]
```

```python
# Split the address column
# View if the changes are correct as required before commit
# And by default it creates dumpy columns
customer_df.Address.str.split(",",expand = True)
```

```python
# Accessing details of a customer at 2nd position to see if there is a full address
customer_df.iloc[2]
```

```python
#  Adding new columns by splitting the 'Address' to make the 'Street_name','Location','Town/City'
# Make changes permanent
customer_df[['Street_name','Location','Town/City']] = customer_df.Address.str.split(",",expand = True)
```

```python
# View and check if the new columns are present
# And a sample of about 50% is viewed
customer_df.sample(frac = 0.5)
```

```python
# Capitalize the new columns 
# Make the changes permanent

customer_df['Street_name'] = customer_df['Street_name'].str.capitalize()
customer_df['Town/City'] = customer_df['Town/City'].apply(lambda x: str(x).capitalize())
customer_df['Location'] = [str(x).capitalize() for x in customer_df['Location']]
```

```python
# Drop the "Address" column
# And save the changes
customer_df.drop(columns = ['Address'],inplace=True)
```

```python
# View the "Date" column and check if it's in good format 
customer_df.Date
```

```python
# Change the data type of 'Date' to datetime
# Specify the format and handle errors
customer_df.Date = pd.to_datetime(customer_df.Date,format = '%d-%m-%Y',errors = 'coerce')
```

```python
# Check the unique values of 'Payment_method'
customer_df.Payment_method.unique()
```

```python
# Capitalize the Payment method
# And view if the changes are correct as required before commit
[x.capitalize() for x in customer_df.Payment_method]
```

```python
# make the changes permanent
customer_df.Payment_method = [x.capitalize() for x in customer_df.Payment_method]
```
